```python
df = pd.read_csv(file)
df = df[df['Grade'] != 'All Grades']
df['Grade'] = df['Grade'].astype('int64')
df.info()
print('shape: ' + str(df.shape[0]) + ' rows ' + str(df.shape[1]) + ' columns.')
```

```python
df_grouped = df.groupby(['Grade', 'Year', 'Category']).agg('sum')
df_grouped
```

```python
df_grade_grouped = df.groupby('Grade')[['Mean Scale Score']].agg('mean')
df_grade_grouped
```

```python
fig, (ax, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
sns.set()
sns.lineplot(data=df_grouped, x='Year', y='Mean Scale Score', hue='Grade', ci=False, ax=ax, linewidth=4)
sns.lineplot(data=df_grouped, x='Year', y='Mean Scale Score', hue='Category', ci=False, ax=ax2, linewidth=4)
ax.set_title('Mean Scale Score vs. Year by Grade')
ax2.set_title('Mean Scale Score vs. Year by Ethnicity')
```

```python
fig2, pl1 = plt.subplots(figsize=(12, 6))
sns.lineplot(data=df_grade_grouped, x=df_grade_grouped.index, y='Mean Scale Score', ax=pl1, linewidth=4)
pl1.set_title('Mean Scale Score by Grade')
```
